```python
def pengisian_bahan_bakar_minimum(SPBU_posisi, jarak_tempuh_maks):
    posisi_saat_ini = 0
    jumlah_isi = 0
    N = len(SPBU_posisi)
    
    try:
        S_INDEX = SPBU_posisi.index(posisi_saat_ini)
    except ValueError:
        return -1 
    
    while posisi_saat_ini < 1000:
        batas_jangkauan = posisi_saat_ini + jarak_tempuh_maks
        if batas_jangkauan >= 1000:
            break
        SPBU_terjauh_index = -1
        p = S_INDEX + 1
        while (p < N) and (SPBU_posisi[p] <= batas_jangkauan):
            SPBU_terjauh_index = p
            p += 1
        if SPBU_terjauh_index == -1:
            return -1
        if SPBU_terjauh_index > S_INDEX:
            posisi_saat_ini = SPBU_posisi[SPBU_terjauh_index]
            jumlah_isi += 1
            S_INDEX = SPBU_terjauh_index
        else:
            return -1
    return jumlah_isi

SPBU_posisi_contoh = [0, 50, 120, 250, 380, 500, 630, 750, 850, 920, 980, 1000]

jarak_1 = 150
hasil_1 = pengisian_bahan_bakar_minimum(SPBU_posisi_contoh, jarak_1)
print(f"Jarak Max: {jarak_1}, Jumlah Isi: {hasil_1}")

jarak_2 = 250
hasil_2 = pengisian_bahan_bakar_minimum(SPBU_posisi_contoh, jarak_2)
print(f"Jarak Max: {jarak_2}, Jumlah Isi: {hasil_2}")

jarak_3 = 15
hasil_3 = pengisian_bahan_bakar_minimum(SPBU_posisi_contoh, jarak_3)
print(f"Jarak Max: {jarak_3}, Jumlah Isi: {hasil_3}")
```
